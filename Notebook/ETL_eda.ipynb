```python
# remplazo los nulos  por el valor correcto 
df['Postal_Code'] = df['Postal_Code'].fillna(5401)
```

```python
# verifico
df[(df['City'] == 'Burlington') & (df['State'] == 'Vermont')].head()
```

```python
df.info()
```

```python
#  Postal_Code a numero entero 
df['Postal_Code'] = df['Postal_Code'].astype(int)
```

```python
# Convertir las columnas 'Order_Date' y 'Ship_Date' de tipo object a tipo fecha con formato "día mes año"
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d/%m/%Y')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='%d/%m/%Y')
```

```python
df.info()
```

```python
# Exporto a un nuevo csv

df.to_csv('superstore_etl.csv', index=False)
```

```python
# Extrae el mes y el año de la columna 'Order_Date'
df['year'] = df['Order_Date'].dt.year
df['month'] = df['Order_Date'].dt.month

# Agrupa las ventas por mes y año y calcula la suma de ventas para cada combinación de mes y año
ventas_por_mes_y_año = df.groupby(['year', 'month'])['Sales'].sum()

# Grafica las ventas por mes y año con líneas continuas
fig, ax = plt.subplots(figsize=(12, 6))
ventas_por_mes_y_año.plot(marker='o', linestyle='-', ax=ax)
plt.xlabel('Año - Mes')
plt.ylabel('Ventas')
plt.title('Ventas por Mes y Año')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión para el campo 'Ventas'
plt.figure(figsize=(8, 6))
plt.scatter(range(len(df)), df['Sales'], marker='o', color='b', label='Datos')
outliers_precio = df[df['Sales'] > df['Sales'].quantile(0.95)]
plt.scatter(outliers_precio.index, outliers_precio['Sales'], marker='x', color='r', label='Outliers')
plt.title('Gráfico de Dispersión - Precio de venta')
plt.xlabel('Índice')
plt.ylabel('Precio de venta')
plt.legend()
plt.show()
```
